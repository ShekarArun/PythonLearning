# Q-Q Plots

Used to check if 2 distributions are of the same type<br>
Take 100 percentiles of the sorted distribution and plot against each other. If in a straight line, then type of distribution is same.

```python
#Import packages
import numpy as np
import scipy.stats as stats
import pylab
```

```python
#Normal distribution
norm_dist = np.random.normal(loc = 0, scale = 1, size = 1000)


# #Find percentiles
# for i in range(0,101):
#     print(i, np.percentile(norm_dist,i))
```

```python
#Generate another normal distribution with mean 25 and variance 4
dist2 = np.random.normal(loc = 25, scale = 4, size = 1000000) #Vary size and see differences

#Plot Q-Q
stats.probplot(dist2, dist = 'norm', plot = pylab)
pylab.show()
```

```python
#Try with a different type of distribution
dist3 = np.random.uniform(low = -1, high = 1, size = 100000)

stats.probplot(dist3, dist = 'norm', plot = pylab)
pylab.show()
```
